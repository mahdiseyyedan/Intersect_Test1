```python
df = data

# Keep only the first row for each ID
df_first = df.drop_duplicates(subset='ID', keep='first')
df_first
```

```python
distances = [    0,    1.818,    3.636,    5.455,    7.273,    9.091,    10.91,    12.73,    14.55,    16.36,    18.18,    20,    21.82,    23.64,    25.45,    27.27,    29.09,    30.91,    32.73,    34.55,    36.36,    38.18,    40]
```

```python
# Repeat each row len(distances) times
df_repeated = df_first.loc[df_first.index.repeat(len(distances))].reset_index(drop=True)

df_repeated['Distance from the Center Line (m)'] = distances * len(df_first)
df_repeated.head(len(distances)+2)
```

```python
df_repeated = df_repeated.drop('Max Lateral Wall Deflection (δw)', axis=1)
df_repeated = df_repeated.drop('Depth (D) [m]', axis=1)
df_repeated = df_repeated.drop('Lateral Wall Deflection (δw) [cm]', axis=1)
df_repeated
```

```python
# Specify your new Excel file path
new_excel_file_path = 'Empty_Heave_TEST.xlsx'

# Save the repeated DataFrame to a new Excel file
df_repeated.to_excel(new_excel_file_path, index=False)
```
